```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
players = pd.read_csv("players.csv")
```

```python
del players["Unnamed: 0"]
```

```python
players["Player"] = players["Player"].str.replace("*", "",regex=False)
```

```python
def single_row(df):
    if df.shape[0] ==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
del players["Rk"]
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps, how = "outer", on=["Player","Year"])
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won","Pts Max","Share"]] = combined[["Pts Won","Pts Max","Share"]].fillna(0)
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(30)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.tail(50)
```

```python
teams["Team"] = teams["Team"].str.replace("*","",regex=False)
teams["Team"] = teams["Team"].str.replace("\xa0","",regex=False)
loop_list = map(str, range(15+1))
print(loop_list)
for ele in loop_list:
    teams["Team"] = teams["Team"].str.replace(str(ele),"",regex=False)
teams["Team"] = teams["Team"].str.replace("()","",regex=False)
teams["Team"] = teams["Team"].str.replace("Philadelphia ers","Philadelphia 76ers",regex=False)
```

```python
teams.tail(30).sort_values("L",ascending=True)
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
display(nicknames)
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```
